```python
import pandas as pd
import numpy as np
```

```python
df= pd.DataFrame([[8,8,4],[7,9,5],[6,7,6],[5,12,7]],columns=['cgpa','profile_score','lpa'])
df
```

```python
def initialize_parameters(layer_dims):

  np.random.seed(3)
  parameter={}

  L= len(layer_dims)

  for l in range (1,L):

    parameter['W'+str(l)] = parameter['W'+str(l)] = np.random.randn(layer_dims[l-1], layer_dims[l]) * 0.01
    parameter['b'+str(l)]=np.zeros((layer_dims[l],1))

  return parameter


```

```python
initialize_parameters([2,2,1])
```

```python
def linear_foreward(A_prev, w , b):

   Z = np.dot(w.T, A_prev) + b
   A = 1/(1+np.exp(-Z))   # sigmoid

   return A
```

```python
def L_layer_foreward( X, parameter):

  A= X

  L= len(parameter) // 2

  for l in range(1, L + 1):
    A_prev= A

    w1= parameter['W'+ str(l)]
    b1= parameter['b' + str(l)]

    # print("A" + str(l)+ " :" , A_prev)
    # print('w' + str(l)+ " :",  w1 )
    # print( 'b' + str(l) + " :", b1)

    A = linear_foreward(A_prev, w1, b1)
    print("A"+ str(l)+ ": ", A)

  return A , A_prev
```

```python
####-------- Epoch Implimentation ------\

parameters= initialize_parameters([2,2,1]) # Initialize parameters once before the loop

epochs= 10

for i in range(epochs):

  Loss= []

  for j in range( df.shape[0]):

   x= df[['cgpa', 'profile_score']].values[j].reshape(2,1)
   y= df[['lpa']].values[j][0]

   # Pass the correct 'parameters' variable to L_layer_foreward
   y_hat, A1 = L_layer_foreward(x, parameters)

   y_hat= y_hat.item()

   # Pass the correct 'parameters' variable to update_parameters
   update_parameters(parameters, x, y, A1, y_hat)

   Loss.append((y- y_hat)**2)

  print("Epoch -", i +1 , " Loss", np.array(Loss).mean())
```
